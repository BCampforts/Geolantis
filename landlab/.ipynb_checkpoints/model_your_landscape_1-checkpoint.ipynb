```python
# Import packages
from landlab import RasterModelGrid, imshow_grid, imshowhs_grid
import numpy as np


mg = RasterModelGrid((50,50), 100)
z = mg.add_ones('topographic__elevation', at = 'node')
z[mg.core_nodes]+= np.random.rand(len(z[mg.core_nodes]))

imshow_grid(mg, 'topographic__elevation')

```

Can you make a grid of 20 by 20 nodes? 

```python
from landlab.components import FastscapeEroder, LinearDiffuser, FlowDirectorD8,FlowAccumulator, FlowDirectorSteepest



mg = RasterModelGrid((20,20), 100)
z = mg.add_ones('topographic__elevation', at = 'node')
z+= np.random.rand(len(z))

# FR = PriorityFloodFlowRouter(mg, update_flow_depressions=True, flow_metric='D4')
FA = FlowAccumulator(mg, flow_director="D8")

K_sp= 0.001
m= .5
n=1

fluvial_incision = FastscapeEroder(mg, K_sp = K_sp, m_sp=m ,n_sp =n)
linear_diffusion = LinearDiffuser(mg, linear_diffusivity=0.01)

time = 0
total_time = 2e5
dt = 1
U = 0.001
while time<total_time: 
    time += dt
    mg.at_node['topographic__elevation'][mg.core_nodes] += U*dt
    
    FA.run_one_step()
    fluvial_incision.run_one_step(dt = dt)
#     linear_diffusion.run_one_step(dt = 100)
```

```python
imshowhs_grid(mg, 'topographic__elevation',altdeg = 1)    
```

Are we in a steady state? 

```python
import matplotlib.pyplot as plt 
plt.figure()
plt.scatter(mg.at_node['drainage_area'], mg.at_node['topographic__steepest_slope'])
plt.xscale('log')
plt.yscale('log')

S_an = np.power((U/(K_sp*np.power(mg.at_node['drainage_area'][mg.core_nodes],m))),1/n)
plt.plot(mg.at_node['drainage_area'][mg.core_nodes],S_an)
plt.show()
```

Let's scale up the problem

Build a grid of 50 by 100 cells. Use timesteps of 200 year. and run the model for 1e6 years. 

```python
from landlab.components import FastscapeEroder, LinearDiffuser, FlowDirectorD8,FlowAccumulator, FlowDirectorSteepest



mg = RasterModelGrid((100,200), 100)
z = mg.add_ones('topographic__elevation', at = 'node')
z+= np.random.rand(len(z))

# FR = PriorityFloodFlowRouter(mg, update_flow_depressions=True, flow_metric='D4')
FA = FlowAccumulator(mg, flow_director="D8")

K_sp= 0.001
m= .5
n=1

fluvial_incision = FastscapeEroder(mg, K_sp = K_sp, m_sp=m ,n_sp =n)
linear_diffusion = LinearDiffuser(mg, linear_diffusivity=0.01)

time = 0
total_time = 1e6
dt = 200
U = 0.001
while time<total_time: 
    time += dt
    mg.at_node['topographic__elevation'][mg.core_nodes] += U*dt
    
    FA.run_one_step()
    fluvial_incision.run_one_step(dt = dt)
#     linear_diffusion.run_one_step(dt = 100)
```

```python
imshowhs_grid(mg, 'topographic__elevation',altdeg = 1)    
```
